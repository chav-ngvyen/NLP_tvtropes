```python
def trope_scraper(url = None):
    
    # Download the webpage
    page = requests.get(url)

    # If a connection was reached
    if page.status_code == 200:

        # Parse
        soup = BeautifulSoup(page.content, 'html.parser')
        
        # Main content
        content = []
        
        # Pick paragraphs with links to other paragraphs
        for i in soup.select("div > p > a[href^='/pmwiki/pmwiki.php/Main/']", {"class":"twikilink"}):
            para = i.parent.text
            content.append(para)
        
        # Join the paragraphs
        description = " ".join(unique(content))
        
        # Meta data
        full_url = url
        short_url = url.split("/")[-1]
        trope_title = soup.find("meta",  property="og:title").get("content") 
        trope_title = trope_title.split(" - ")[0]
        
        # Related content
        related_trope = []
        related_description = []

        for i in soup.select("div > p > a[href^='/pmwiki/pmwiki.php/Main/']", {"class":"twikilink"}):
            # Get the last part of the URL
            related_url = i['href']
            
            related_trope.append(i['href'].split("/")[-1])
            related_description.append(i.text)
#         print(related_url)
#         print(related_trope)
        my_dict = dict(zip(related_trope, related_description))
#         related_dict = dict(zip(related_url, related_trope))
        return [trope_title, short_url, description, my_dict,full_url]
```

```python
def multi_trope_scraper(urls = None, sleep = 7):
    scraped_data = []
    for url in urls:
        
        print(url)
        
        try:
            # Scrape the content This will break on URLs that we haven't
            # accounted for the structure on. So we'll use a try and except 
            # clause so our code continues even though it breaks on some urls. 
            scraped_data.append(trope_scraper(url))
        except:
            print("URL doesn't work with scraper")

        # Put the system to sleep for a random draw of time (be kind)
        time.sleep(random.uniform(0,sleep))

    dat = pd.DataFrame(scraped_data,columns=["trope_name","short_url","description","related","url"])
    return dat
```
