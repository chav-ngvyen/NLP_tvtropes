```python
import pandas as pd
import numpy as np

import pandasql as ps
from pandasql import sqldf

import matplotlib.pyplot as plt
import seaborn as sns

# Ignore matplotlib warnings
# import warnings
# warnings.filterwarnings("ignore")
```

```python
tv = pd.read_csv("../TVTropesData/tv_tropes.csv")
tv["media"] = "tv"
```

```python
film = pd.read_csv("../TVTropesData/film_tropes.csv")
film["media"] = "film"
```

```python
lit = pd.read_csv("../TVTropesData/lit_tropes.csv")
lit["media"] = "lit"
```

```python
df = pd.concat([tv, film, lit])
df.head(5)
```

For each trope, count unique appearance across 3 mediums

```python
# Count
q = ("""
    SELECT DISTINCT Trope, media, COUNT(Title) as count
    FROM df
    GROUP BY Trope, media
    ORDER BY count desc;
    """)
top = sqldf(q)
```

```python
# Count number of tropes for each title
q = ("""
    SELECT DISTINCT Title, media, COUNT(Trope) as count
    FROM df
    GROUP BY Title, media
    ORDER BY count desc;
    """)
tot = sqldf(q)
```

```python
# Count number of tropes for each title
top_tv = ("""
    SELECT DISTINCT Trope, COUNT(Title) as count
    FROM df
    WHERE media ='tv'
    GROUP BY Trope
    ORDER BY count desc
    LIMIT 10;
    """)
top_tv = sqldf(top_tv)
```
